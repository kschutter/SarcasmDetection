```python
import numpy as np
import pandas as pd
from pandasql import sqldf
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv('../data/train-balanced-sarcasm.csv')
```

```python
all_sarcasm = """
                Select comment
                FROM df
                WHERE label=1
                """

sarcastic_df = sqldf(all_sarcasm, locals())
```

```python
sarcastic_df.head()
```

```python
text = ""
for row in sarcastic_df.iterrows():
    text += " " + str(row[1])
```

```python
text = text.replace("\n", "")
text = text.replace("comment", "")
text = text.replace("dtype: object", "")
text = text.replace("Name", "")
```

```python
stopwords = set(STOPWORDS)
# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords, max_words=100, background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
